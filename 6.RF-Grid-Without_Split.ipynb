```python
import pandas as pd
```

```python
dataset=pd.read_csv("50stratup.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
independent
```

```python
dependent=dataset[['Profit']]
dependent
```

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
independent=sd.fit_transform(independent)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
param_grid={'n_estimators':[10,50,100],'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],'max_features':['sqrt', 'log2']}
```

```python
grid=GridSearchCV(RandomForestRegressor(), param_grid, refit=True, verbose=3,n_jobs=-1)
```

```python
grid.fit(independent,dependent)
```

```python
#print best parameter after tuning
result=grid.cv_results_
print("Thte R_Score value for best parameter:{}".format(grid.best_params_))
```

```python
#Table
table=pd.DataFrame.from_dict(result)
table
```

```python
y_predt=grid.predict(independent)
from sklearn.metrics import r2_score
r_score=r2_score(dependent,y_predt)
r_score
```
