```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
independent=sd.fit_transform(independent)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sd=StandardScaler()
independent=sd.fit_transform(independent)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LinearRegression
```

```python
param_grid={'fit_intercept':[True,False],'copy_X':[True,False],'positive':[False,True]}
```

```python
grid=GridSearchCV(LinearRegression(), param_grid, refit=True, verbose=3,n_jobs=-1)
```

```python
grid.fit(independent,dependent)
```

```python
#print best parameter after tuning
result=grid.cv_results_
print("Thte R_Score value for best parameter:{}".format(grid.best_params_))
```

```python
#Table
table=pd.DataFrame.from_dict(result)
table
```

```python
y_predt=grid.predict(independent)
from sklearn.metrics import r2_score
r_score=r2_score(dependent,y_predt)
r_score
```
