```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sd=StandardScaler()
```

```python
X_train=sd.fit_transform(X_train)
X_train
```

```python
X_test=sd.transform(X_test)
X_test
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.svm import SVR
```

```python
param_grid={'kernel':['linear', 'poly', 'rbf', 'sigmoid'],'C':[10,100,1000,2000,3000,5000],'gamma':['auto','scale']}
```

```python
grid=GridSearchCV(SVR(), param_grid, refit=True, verbose=3,n_jobs=-1)
```

```python
grid.fit(X_train,Y_train)
```

```python
#print best parameter after tuning
result=grid.cv_results_
print("Thte R_Score value for best parameter:{}".format(grid.best_params_))
```

```python
#Table
table=pd.DataFrame.from_dict(result)
table
```

```python
y_predt=grid.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_predt)
r_score
```
